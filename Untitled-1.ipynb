```python
import pandas as pd
import numpy as np

import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.stats.api as sms
```

```python
df = pd.read_parquet('data_enigh_2.parquet')
df = df.loc[df.estado=='Ciudad de México',:]
df.head()
```

```python
#Eliminamos las columnas que actualmente son de tio string, ya que las volveremos dummies más adelante
df = df.drop(columns=['estado', 'municipio'])
```

```python
model = smf.ols(formula='renta ~ antiguedad + num_cuarto + bano_comp + focos_ahor +  tot_hom + tot_muj', data=df)

results = model.fit()

print(results.summary())
```

```python
X = df[['antiguedad', 'num_cuarto', 'bano_comp', 'focos_ahor', 'tot_hom', 'tot_muj']]

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = pd.DataFrame()
vif_data['Variable'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

vif_data
```

```python
# Aplicamos una transformación logartimica al valor Y: Renta
df['renta_log'] = np.log(df['renta'])
formula='renta_log ~ antiguedad + num_cuarto + bano_comp + focos_ahor +  tot_hom + tot_muj'
model = smf.ols(formula=formula, data=df).fit()

print(model.summary())
```

```python
fitted_values = model.fittedvalues
residuals = model.resid
plt.figure(figsize=(8, 5))
plt.scatter(fitted_values, residuals, alpha=0.5)
plt.axhline(0, color='red', linestyle='--', linewidth=2)
plt.xlabel('Valores Predichos')
plt.ylabel('Residuos')
plt.title('Gráfico de Residuos vs. Valores Predichos')
plt.show()
```

```python
residuals = model.resid

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(residuals, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histograma de Residuos')
plt.xlabel('Residuos')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
sm.qqplot(residuals, line='s', ax=plt.gca())
plt.title('Gráfico Q-Q de Residuos')

plt.tight_layout()
plt.show()
```

```python
df_2 = pd.read_parquet('data_enigh_2.parquet')
df_2 = df_2.loc[df_2.estado=='Ciudad de México',:]
df_2['renta_log'] = np.log(df_2['renta'])
df_2 = df_2.drop(columns=['tot_resid','renta','estado', 'cuart_dorm'])
df_2.head(5)
```

```python
df_numerico = pd.get_dummies(df_2, columns=['municipio'], drop_first=True)
df_numerico.columns = df_numerico.columns.str.replace(' ', '_')
df_numerico = df_numerico.rename(columns={'municipio_Gustavo_A._Madero': 'municipio_Gustavo_A_Madero'})
df_numerico.info()
```

```python
formula = 'renta_log ~ ' + ' + '.join(df_numerico.drop(columns=['renta_log']).columns)

# Crea el modelo de regresión
model = smf.ols(formula=formula, data=df_numerico).fit()

# Imprime el resumen del modelo
print(model.summary())
```

```python
def histogram_plot(df, column):
    plt.figure(figsize=(10, 6))
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black', alpha=0.7)

    plt.title(f'Histograma {column}', fontsize=16, weight='bold')
    plt.xlabel('Valor', fontsize=14)
    plt.ylabel('Frecuencia', fontsize=14)

    plt.grid(True, linestyle='--', alpha=0.6)

    plt.axvline(df[column].mean(), color='red', linestyle='--', linewidth=2, label=f"Media: {df[column].mean():.2f}")
    plt.axvline(df[column].mean() + df[column].std(), color='black', linestyle=':', linewidth=2, label=f"+1 Desviación Estándar: {df[column].mean() + df[column].std():.2f}")
    plt.axvline(df[column].mean() + 2*df[column].std(), color='orange', linestyle=':', linewidth=2, label=f"+2 Desviaciones Estándar: {df[column].mean() + 2*df[column].std():.2f}")
    plt.axvline(df[column].mean() - df[column].std(), color='black', linestyle=':', linewidth=2, label=f"-1 Desviación Estándar: {df[column].mean() - df[column].std():.2f}")
    plt.axvline(df[column].mean() - 2*df[column].std(), color='orange', linestyle=':', linewidth=2, label=f"-2 Desviaciones Estándar: {df[column].mean() - 2*df[column].std():.2f}")

    plt.legend(fontsize=12)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    plt.show()
```

```python
histogram_plot(df, 'renta_log')
```
