```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
def histogram_plot(df, column):
    plt.figure(figsize=(10, 6))
    plt.hist(df[column], bins=30, color='skyblue', edgecolor='black', alpha=0.7)

    plt.title(f'Histograma {column}', fontsize=16, weight='bold')
    plt.xlabel('Valor', fontsize=14)
    plt.ylabel('Frecuencia', fontsize=14)

    plt.grid(True, linestyle='--', alpha=0.6)

    plt.axvline(df[column].mean(), color='red', linestyle='--', linewidth=2, label=f"Media: {df[column].mean():.2f}")
    plt.axvline(df[column].mean() + df[column].std(), color='black', linestyle=':', linewidth=2, label=f"+1 Desviación Estándar: {df[column].mean() + df[column].std():.2f}")
    plt.axvline(df[column].mean() + 2*df[column].std(), color='orange', linestyle=':', linewidth=2, label=f"+2 Desviaciones Estándar: {df[column].mean() + 2*df[column].std():.2f}")
    plt.axvline(df[column].mean() - df[column].std(), color='black', linestyle=':', linewidth=2, label=f"-1 Desviación Estándar: {df[column].mean() - df[column].std():.2f}")
    plt.axvline(df[column].mean() - 2*df[column].std(), color='orange', linestyle=':', linewidth=2, label=f"-2 Desviaciones Estándar: {df[column].mean() - 2*df[column].std():.2f}")

    plt.legend(fontsize=12)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    plt.show()
```

```python
def box_plot(df,column):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=column, data=df, color='skyblue', width=0.4)

    plt.title(f'Distribución de la {column}', fontsize=16, weight='bold')
    plt.xlabel(f'{column}', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.6)

    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    plt.show()
```

### modelo


```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df_numerico = df.drop(columns=['estado', 'municipio'])
```

```python
model = smf.ols(formula='renta ~ antiguedad + cuart_dorm + num_cuarto + bano_comp + focos_ahor + tot_resid + tot_hom + tot_muj', data=df_numerico)

results = model.fit()

print(results.summary())
```

### multicolinealidad

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df_numerico = df.drop(columns=['estado', 'municipio'])
matriz_corr = df_numerico.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(matriz_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, square=True)

plt.title('Matriz de Correlación', fontsize=18, weight='bold')
plt.xticks(fontsize=8, rotation=45, ha='right') 
plt.yticks(fontsize=8, rotation=45, ha='right')

plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.pairplot(df_numerico, diag_kind='kde', palette='coolwarm')

plt.show()
```

```python
X = df[['antiguedad', 'cuart_dorm', 'num_cuarto', 'bano_comp', 'focos_ahor', 'tot_resid', 'tot_hom', 'tot_muj']]


vif_data = pd.DataFrame()
vif_data['Variable'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

vif_data
```

### Heterocedasticidad

```python
from statsmodels.stats.diagnostic import het_breuschpagan
```

```python
fitted_values = results.fittedvalues
residuals = results.resid
plt.figure(figsize=(8, 5))
plt.scatter(fitted_values, residuals, alpha=0.5)
plt.axhline(0, color='red', linestyle='--', linewidth=2)
plt.xlabel('Valores Predichos')
plt.ylabel('Residuos')
plt.title('Gráfico de Residuos vs. Valores Predichos')
plt.show()
```

```python
residuals = results.resid
exog = results.model.exog

bp_test = het_breuschpagan(residuals, exog)

labels = ['LM Statistic', 'p-value', 'F-Statistic', 'F p-value']
bp_results = dict(zip(labels, bp_test))

print('Breusch-Pagan test results:')
for key, value in bp_results.items():
    print(f'{key}: {value}')
```

### No normalidad de los errores

```python
from statsmodels.stats.stattools import jarque_bera
```

```python
residuals = results.resid

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(residuals, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histograma de Residuos')
plt.xlabel('Residuos')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
sm.qqplot(residuals, line='s', ax=plt.gca())
plt.title('Gráfico Q-Q de Residuos')

plt.tight_layout()
plt.show()
```

```python
residuals = results.resid

jb_test = jarque_bera(residuals)

jb_statistic, jb_pvalue, skewness, kurtosis = jb_test

print('Jarque-Bera test results:')
print(f'Estadístico Jarque-Bera: {jb_statistic}')
print(f'p-valor: {jb_pvalue}')
print(f'Skewness (Asimetría): {skewness}')
print(f'Kurtosis (Curtosis): {kurtosis}')
```
